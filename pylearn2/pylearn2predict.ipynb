```python
from pylearn2.utils import serial

from theano import tensor as T
from theano import function

import numpy as np
import sys
import csv

from pylearn2.datasets.dataclassraw import CLICK9DAYS
from pylearn2.datasets.transformer import TransformerClick
```

### Dataset and Model

```python
model = serial.load('LH1_1.pkl')
```

```python
rawdata = CLICK9DAYS(which_set='valid')
transformer = TransformerClick(raw=rawdata, interaction=False, nfeatures=23)
```

```python
batch_size = 512
model.set_batch_size(batch_size)
```

```python
X = model.get_input_space().make_batch_theano()
Y = model.fprop(X)
```

```python
#Y = T.argmax(Y, axis=1)
```

```python
f = function([X], Y)
```

### Get probabilities of all classes

```python
y = []
```

```python
batches = rawdata.X.shape[0] / batch_size
```

```python
%%time
for i in xrange(batches):
    x_arg = transformer.perform(rawdata.X[i*batch_size:(i+1)*batch_size])
    
    y.append(f(x_arg.astype(X.dtype)))
```

```python
nexamples = rawdata.X.shape[0]
extra = batch_size - nexamples % batch_size
assert (nexamples + extra) % batch_size == 0

if extra > 0:
    x_arg = np.concatenate((transformer.perform(rawdata.X[batches*batch_size:]), np.zeros((extra, 23),
    dtype=X.dtype)), axis=0)
    y.append(f(x_arg.astype(X.dtype)))
```

```python
y
```

```python
y = np.concatenate(y)
#assert y.ndim == 1
```

```python
y = y[:nexamples]
```

### Get the expected class score

```python
ypred = y[:,1]
```

```python
import matplotlib.pyplot as plt
######################################
fig, axes = plt.subplots()
axes.set_xlabel('')
axes.set_ylabel('')
axes.plot(ypred, 'ro')
plt.show()
#fig.savefig(device + "_timestamp.svg")
#######################################
```

```python
def logloss(p, y):
    ''' FUNCTION: Bounded logloss

        INPUT:
            p: our prediction
            y: real answer

        OUTPUT:
            logarithmic loss of p given y
    '''

    p = max(min(p, 1. - 10e-15), 10e-15)
    return -y*log(p) - (1-y)*log(1. - p)
```

### Write prediction to csv

```python
f = open("submission9.csv", "wb")
writer = csv.writer(f, delimiter=',')
writer.writerow(['QuestionId', 'IsTrue'])
for i in xrange(len(ypred)):
    writer.writerow([str(i+1), '{0:.15f}'.format(ypred[i])])
f.close()
```

### information retrival

```python
def fprop(X, model, return_all=False):

    rval = model.layers[0].fprop(X)

    rlist = [rval]
    
    for layer in model.layers[1:-1]:
        rval = layer.fprop(rval)
        rlist.append(rval)

    if return_all:
        return rlist
    return rval
```
